```python
store_data=pd.read_csv("superstore_final_dataset.csv.csv")
store_data.head(5)
store_data.dtypes
```

```python
store_data.describe()
```

```python
# store_data['Order_Date'] = pd.to_datetime(store_data['Order_Date'], errors='coerce')
store_data['Order_Date'] = store_data['Order_Date'].astype(str).str.extract(r'\/(\d+)\/')
store_data=store_data.head(20)
store_data = store_data.dropna(subset=['Order_Date'])
store_data['Order_Date']
```

```python
# store_data['Order_Date'] = pd.to_numeric(store_data['Order_Date'])
grouped_data = store_data.groupby('Sales')['Order_Date'].sum().reset_index()
grouped_data.sum()
```

```python
plt.plot(grouped_data['Order_Date'], grouped_data['Sales'], marker='x', linestyle='--')

# Labels and title
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.title('Sales Trend Over Time')

# Improve x-axis labels
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_data = store_data.groupby('Sales')['Order_Date'].sum().reset_index().sort_values(by='Order_Date')
# grouped_data=grouped_data.head(10000)
fig, ax = plt.subplots()

# plt.bar(grouped_data['Order_Date'], grouped_data['Sales'])
ax.bar(grouped_data['Order_Date'], grouped_data['Sales'], color='skyblue')
# Labels and title
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.title('Sales Trend Over Time')

# Improve x-axis labels
plt.xticks(rotation=90)
plt.show()
```

```python
historical_data=pd.read_csv("historical weather data.csv")
historical_data['dt_iso'] = pd.to_datetime(historical_data['dt_iso'], errors='coerce')
historical_data['dt_iso']
```

```python

averages = []
columns = []

for col in historical_data.columns:
    if pd.api.types.is_numeric_dtype(historical_data[col]):  # Ensure column is numeric
        averages.append(historical_data[col].mean())
        columns.append(col)

# Ensure we have enough colors for the number of columns
colors = ['red', 'blue', 'green', 'pink', 'darkblue', 'purple', 'skyblue', 'black', 'brown']
colors = colors[:len(columns)]  # Adjust colors list if needed

plt.bar(columns, averages)  
plt.xlabel("Weather Metrics")
plt.ylabel("Averages")
plt.title("Temperature Averages Against Weather Metrics")
plt.grid(True)
plt.xticks(rotation=80)  # Rotate labels for better readability
plt.show()
```
